### Advent of Transfer Learning
* Let's use existing trained models for our issues
  * DL 2.0
* Transformers are now becoming the general purpose architecture and replacing traditional methods

### Current style of Transfer Learning MLOps

1. Find the task that matches your problem
  * text: sentiment, summarization
  * Images: classification, object detection, separate background
  * Speech: speec to text, classification, translation



2. Pick a pre-trained state of art model
   * No need to build, label, and clean a large dataset
   * 2 lines of code only to download and test a model
  

3. If needed, fine-tune model for use case
